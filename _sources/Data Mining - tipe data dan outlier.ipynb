# Data Understanding - Tipe data dan Deteksi Outlier

## Tipe Data dalam Data Mining



Tipe data adalah sebuah konsep dalam pemrograman yang digunakan untuk mengklasifikasikan dan menentukan jenis nilai yang dapat digunakan dalam suatu program. Setiap tipe data memiliki karakteristik unik yang mempengaruhi cara penyimpanan di memori serta operasi yang bisa dilakukan terhadapnya. Dalam Data Mining tipe datanya terdiri dari Kategorikal dan Numerik. Dari masing-masing tipe data tNomindibagi menjadi beberapa tipe data yakni Nominal, Biner, Ordinal, Interval, dan Rasio.

- **Tipe Data Kategorikal**

  Merupakan jenis data yang merepresentasikan kategori atau label. Data ini sering digunakan dalam klasifikasi, clustering, dan analisis statistik lainnya.Tipe data Kategorikal terdiri dari Nominal, Biner, Ordinal. Terdapat ciri-ciri data Kategorikal seperti:
  *   Tidak bisa dihitung atau dioperasikan secara matematis seperti angka.
  *   Bisa berbentuk teks atau angka yang hanya berfungsi sebagai label.

1. Tipe Data Nominal

  Tipe Data Nominal merupakan tipe data yang paling sederhana dimana nilai yang diberikan atau record hanya berupa label atau sekedar penamaan saja, tidak memiliki ukuran dan ukuran. Contoh tipe data ini seperti nama, kelas, teks, tanggal, waktu, mata uang. Meskipun tidak ada ukuran ataupun urutan, kita perlu untuk menghitug frekuensi dari data nominal tersebut. Ukuran yang digunakan untuk data ini adalah modus.

2. Tipe Data Biner

  Tipe data Biner hanya menggunakan 2 kategori. data nominal dapat dituliskan secara nominal yakni "ya" atau "tidak". Representasi angka dalam tipe data ini adalah 1 dan 0. Pada data biner, angka 1 merujuk pada ya dan angka 0 berarti tidak. Seperti data nominal kita menggunkaan frekuensi dan modus dari data Biner.

3. Tipe Data Ordinal
  Data Ordinal merupakan suatu data yang dapat diurutkan dari yang paling rendah sampai yang paling tinggi. Namun, jarak antara tingkatan tidak dapat diukur dan dibandingkan secara numerik. Selain modus, ukuran statistik median dapat digunakan untuk mencari nilai tengah di dalam data.

- **Tipe data Numerik**

  Merupakan tipe data yang digunkan untuk data yang dapat diukur secara kuantitatif, sehingga dapat menerapkan operasi matematika. Tipe data ini terdiri dari Interval dan Rasio.

1. Tipe Data Interval
 Tipe data Interval memiliki jarak tiap tingkatan sehingga operasi matematika yang dapat digunakan yakni penjumlahan dan pengurangan. Namun, nilai nol belum didefinisikan pada data Interval sehinga operasi perkalian dan pembagian belum bisa untuk digunakan. Modus, median, rata-rata sudah dapat digunakan untuk data ini.

2. Tipe Data Rasio
 Tipe data Rasio merupakan data numerik yang lengkap dengan nilai nol. jadi, untuk operasi matematika seperti penjumlahan, pengurangan, perjalian, dan pembagian sudah bisa digunakan dalam tipe data ini.








## Deteksi Outlier Menggunakan KNN

### KNN

**K-Nearest Neighbors (KNN)** merupakan sebuah cara untuk mengklasifikasikan dengan cara melihat sesuatu yang berada di dekatnya. KNN juga disebut dengan Algoritna oembelajar malas.  Karena tidak memerlukan pembelajaran terlebih dahulu, dan langsung meyimpan data set dan pada saat klaifikasi melakukan set data. KNN bekerja dengan menggunakan kedekatan dan pemungutan suara mayoritas untuk membuat prediksi atau *forecasting*. Pada KNN terdapat istilah "k". "k" merupakan angka yang memberi tahu algoritma beberapa banya titik terdekat (tetangga) yang digunakan untuk membuat keputusan. Contohnya:
Misalkan menntukan nama buah dan membandingkannya dengan buah yang telah dikenal. kemudian menggunakan nilai "k" sebanyak 3 kemudian 2 dari 3 merupakan buah apel dan 1 dari 3 merupakan buah pisang. jadi algoritma ini mengatakan bahwa buah tersebut merupakan apel karena sebagian besar tetangganya apel.


 ### Cara Memilih Nilai "k" untuk Algoritma KNN

Dalam KNN, pemilihan nilai "k" sangat penting untuk menentukan hipotesis hasil dari prediksi. Jika kumpulan data memiliki outlier atau noise yang signifikan, nilai "k" yang lebih tinggi dapat membantu memperhalus prediksi dan mengurangi data yang noise. Namun, pemilihan nilai yang tinggi dapat menyebabkan underfitting.

Terdapat metode untuk memilih nilai k:
1. Validasi Silang
  Merupakan metode yang kuat untuk memilih nilai k. metode ini melibatkan pemisahan data menjadi k subswt yang tersisa, kemudian mengulanginya setiap subset. Nilai k yang dihasilkan memiliki akurasi validasi rata-rata tinggi yang merupakan pilihan terbaik.
2. Metode Elbow
  Metode ini memetakan tingkat kesalahan atau akurasi model untuk nilai k yang berbeda. Pada saat meningkatkan nilai "k", tingkat kesalahan akan menurun pada awalnya. Pada titik tertentu, tingkat kesalahan mulai menurun lebih lambat.
3. Nilai Ganjil untuk "K"
  Untuk menentukan hasil, disarankan untuk memilih bilangan ganjil untuk menghindari nilai seri pada saat menentukan kelas dan klasifikasi.

## Mencari Data Outlier dalam Data Iris

### Manfaat Menentukan Data Outlier

1. Meningkatkan Akurasi Analisis Statistik
Outlier bisa menyebabkan distorsi dalam statistik seperti rata-rata dan standar deviasi. Misalnya, dalam perhitungan rata-rata pendapatan jika ada satu orang dengan pendapatan jauh lebih tinggi dibandingkan yang lain, rata-rata akan menjadi bias.

2. Mencegah Kesalahan dalam Model Prediksi (Machine Learning)
Outlier dapat membuat model machine learning menjadi kurang akurat karena model akan belajar dari data yang tidak mewakili pola sebenarnya.

3. Mengidentifikasi Kesalahan Data atau Anomali
Outlier bisa menunjukkan bahwa ada kesalahan dalam pencatatan data atau anomali yang perlu diperiksa lebih lanjut.

4. Membantu Deteksi Fraud atau Kejahatan Finansial
Outlier sering digunakan dalam deteksi penipuan (fraud detection), misalnya dalam transaksi kartu kredit atau transfer bank.

5. Mengoptimalkan Pengambilan Keputusan
Mengetahui outlier dalam data membantu pengambilan keputusan yang lebih baik dengan mempertimbangkan apakah outlier perlu dihapus, dikoreksi, atau dianalisis lebih lanjut.

### Menghitung Jarak dari Menggunakan Encludian Distance

```python
var_merged = merged_db

feature_column = ['petal length', 'petal width', 'sepal length', 'sepal width']
X = var_merged[feature_column]

nilai_k = 5
knn = NearestNeighbors(n_neighbors=nilai_k, metric='euclidean')
knn.fit(X)
distances, _ = knn.kneighbors(X)

df_features = pd.DataFrame(X, columns=feature_column)
df_features['euclidean_distance'] = distances.mean(axis=1)

var_merged['euclidean_distance'] = df_features['euclidean_distance']

print("Hasil dari perhitungan jarak menggunakan Encludian")
print(tabulate(df_features[["sepal length", "sepal width", "petal length", "petal width", "euclidean_distance"]], headers="keys", tablefmt="grid"))
```

**<> Penjelasan Code**
- Tentukan database yang akan dihitung untuk nilai outlier.
- Akses kolom dalam database yang digunakan sebagai fitur.
- Tentukan nilai "k" dan hitung Jarak dari setiap dataset menggunakan metrik Encludian menggunakan library "NearestNeighbors".
- Hitung rata-rata sebaran setiap datasetnya.

### Mencari Data Outlier Menggunakan Metode IQR

```python
Q1 = var_merged['euclidean_distance'].quantile(0.25)
Q3 = var_merged['euclidean_distance'].quantile(0.75)

IQR = Q3 - Q1

# Menentukan batas bawah dan atas
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

# Ambil data outlier
outliers = var_merged[(var_merged['euclidean_distance'] < batas_bawah) |
                      (var_merged['euclidean_distance'] > batas_atas)]
outlier_terbesar = outliers.loc[outliers['euclidean_distance'].idxmax()]

print("Outlier Terbesar:")
print(tabulate(outlier_terbesar.to_frame().T, headers="keys", tablefmt="grid"))
```

**<> Penjelasan Code**
- Gunakan Rumus Interquertile Range untuk mencari data outlier.
- Kemudian tentukan presentil untuk Q1 dan Q2 nya.
- Bandingkan data dengan rumus menentukan batas untuk mengetahui data ouliernya.
